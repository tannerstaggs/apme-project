```python
import pandas as pd

content = open("hurdat2-1851-2023-051124.txt", "r")
hurdat = content.read()
content.close()
```

```python
import re

ATCF_pattern = r"AL\d{6,}"
ATCF_codes = re.findall(ATCF_pattern, hurdat)
print(ATCF_codes)
print(len(ATCF_codes))
```

```python
name_pattern = r"[A-Z-?]{3,}"

names = re.findall(name_pattern, hurdat)
print(names)
print(len(names))
```

```python
fix_pattern = r"[A-Z-?]{3,},\s+(\d{1,3}),"

fixes = re.findall(fix_pattern, hurdat)
print(fixes)
print(len(fixes))

i=0
for fix in fixes: 
    i = i + int(fix)
print(i)
```

```python
date_pattern = r"\d{8}"

dates = re.findall(date_pattern, hurdat)
print(dates)
print(len(dates))
```

```python
UTC_pattern = r"\d{8},\s+(\d{4}),"

UTC = re.findall(UTC_pattern, hurdat)
print(UTC)
print(len(UTC))
```

```python
status_pattern = r"\d{4},\s+,?[A-Z]?,?\s+([A-Z]{2}),"

status = re.findall(status_pattern, hurdat)
print(status)
print(len(status))
```

```python
special_pattern = r"\d{4},\s(\s+|[A-Z]?),"

special_entry = re.findall(special_pattern, hurdat)
print(special_entry)
print(len(special_entry))
```

```python
latitude_pattern = r"\d{0,3}\.\d{1}N"

latitude = re.findall(latitude_pattern, hurdat)
print(latitude)
print(len(latitude))
```

```python
longitude_pattern = r"\d{0,3}\.\d{1}[WE]"

longitude = re.findall(longitude_pattern, hurdat)
print(longitude)
print(len(longitude))
```
